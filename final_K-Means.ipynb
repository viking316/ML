```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Load the data from the CSV file
df = pd.read_csv("Mall_Customers.csv")

X = df.iloc[:, [3, -1]].values
# print(X)

list = []
for i in range(1, 11):
    k = KMeans(n_clusters=i)
    k.fit_predict(X)
    list.append(k.inertia_)

# print(list)

plt.plot(range(1, 11), list)
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.show()

k = KMeans(n_clusters=6)
y = k.fit_predict(X)

plt.scatter(X[y == 0, 0], X[y == 0, 1], label="cluster1")
plt.scatter(X[y == 1, 0], X[y == 1, 1], label="cluster2")
plt.scatter(X[y == 2, 0], X[y == 2, 1], label="cluster3")
plt.scatter(X[y == 3, 0], X[y == 3, 1], label=" cluster4")
plt.scatter(X[y == 4, 0], X[y == 4, 1], label="cluster5")
plt.scatter(X[y == 5, 0], X[y == 5, 1], label="cluster6")

plt.legend()
plt.show()
```
