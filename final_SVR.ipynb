```python
import numpy as np
import pandas as pd
# from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn import *
from sklearn.svm import SVR
import matplotlib.pyplot as plt


ds = pd.read_csv("Salaries.csv")

x = ds.loc[:, ["Level"]].values
y = ds.loc[:, ["Salary"]].values

# STANDARD SCALING THE X AND Y VALUES
sc = StandardScaler()

x = sc.fit_transform(x)
y = sc.fit_transform(y)

# MAKING THE SVR MODEL
svr = SVR(kernel="rbf")
svr.fit(x, y)

# PLOTTING THE OUTPUT
plt.scatter(sc.inverse_transform(x), sc.inverse_transform(y), color="red")
plt.plot(sc.inverse_transform(x), sc.inverse_transform(svr.predict(x).reshape(-1, 1)), color="blue")
plt.title("SVR")
plt.xlabel("LEvel")
plt.ylabel("Salary")
plt.show()


# PLOTTING A SMOOTHER GRAPH
X1 = np.arange(min(sc.inverse_transform(x)), max(sc.inverse_transform(x)), 0.1)
X1 = X1.reshape(len(X1), 1)
plt.scatter(sc.inverse_transform(x), sc.inverse_transform(y), color="red")
plt.plot(X1, sc.inverse_transform(svr.predict(sc.fit_transform(X1)).reshape(-1, 1)), color="blue")
plt.title("SVR")
plt.xlabel("LEvel")
plt.ylabel("Salary")
plt.show()
```
